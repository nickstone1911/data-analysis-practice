# Introduction to Matplotlib
>- Matplotlib is considered the "grandfather" library for data visualization with Python
>- It was designed to replicate MatLab's plotting capabilities in Python
>- Many other popular visualization modules such as `seaborn` and `pandas`'s  visualization are built off of Matplotlib
---
## Major Pros of Matplotlib are:
1. Generally easy to get started for simple plots
2. Support for custom labels and texts
3. Great for control of every element in a figure
4. High-quality output in many formats
5. Highly customizable


---

## Two Major Coding Styles Used with Matplotlib
1. Object Oriented (OO) Style (Focus of this tutorial)
>- Explicitly create Figure and Axes, and call methods on them
>- Offers the most flexibility and customization but comes at a cost of more complexity
2. The function oriented approach with `pyplot`
>- Rely on `pyplot` to automatically create and manage Figures and Axes
>- Uses `pyplot` functions for plotting
>- Often the quickest way to plotting with Matplotlib but without as much flexibility as the OO approach

Which one (OO or pyplot) to use?
>- Depends on your needs
>>- For quick, simple plots, `pyplot` might be the best
>>- For more customization or more complicated graphing requests, you might need to use the OO approach
>- For data science/analytics projects, we often use other modules such as `seaborn` or `pandas` because they offer convenient and still powerful ways to visualize our data. But it is still beneficial to learn the basics of Matplotlib before diving into other visualization modules because most of them are built off of Matplotlib.

---
## Some Definitions
Matplotlib graphs your data on Figure(s) each of which can contain one or more Axes
>- `Figure`: The whole figure. The figure keeps track of all the child `Axes`, a group of "special" Artists (titles, figure legends, colobars, etc), and subfigures
>- `Axes`: An Axes is an Artist attached to a figure that contains a region for plotting data. Each Axes also has a title (set with `set_title()`), an x-label (set with `set_xlabel()`), and a y-label (set with `set_ylabel()`)
>- `Axis`: these objects set the scale, limits, and generate ticks (the marks on the Axis) and ticklabels.
>- `Artist`: Basically, everything visable on the Figure is an *Artist*. When the Figure is rendered, all of the Artists are drawn to the canvas.

# Basic Steps to Generating a Plot With Matplotlib
1. Import Modules
2. Prepare Data
3. Create The Plot
4. Customize The Plot
5. Observe/show the plot

# Step 1: Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Step 2: Prepare Data
>- In this example we use np.linspace() to generate some data
>- np.linspace() returns evenly spaced numbers of a specified interval
>>- np.linspace(*start*, *stop*, *num*) where...
>>>- *start* is the starting value of the sequence
>>>- *stop* is the stopping value of the sequence
>>>- *num* is the number of samples to generate


```python
# Define some data for the tutorial

x = np.linspace(0, 4 * np.pi, 1000)

y = np.sin(x)
```

# Step 3: Create the Plot

1. First, we generate an instance of a `figure.Figure` and `axes.Axes`
>- Think of the figure like a blank canvas
>- Think of the Axes as a part of the canvas on which we make visualizations
>- A common and convenient way to instantiate both the figure and axes objects is with the `subplots()` method
>>- [subplots() doc](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html)
2. Once we have an Axes instance, we plot on top of it

The following is how we create the blank "canvas"
>- We will create a figure with a single axes

```python
# Create a figure and a set of subplots
fig, ax = plt.subplots()
# should show a blank canvas to plot on. This is the axes object
```

Then, we call `plot()` on the `ax` object to generate a plot on our blank canvas

```python
fig, ax = plt.subplots()
# Now, you plot on the axes by calling the plot() method
ax.plot(x,y)

fig.show()
# fig.show() is the equivalent of a print statement
# you don't need this in jupyter notebooks to see the chart but you do in other IDE such as pyCharm
```

# Step 4: Customize the Plot
>- Matplotlib allows us to modify almost everything in a plot
>- See [Matplotlib doc](https://matplotlib.org/stable/index.html) and browse the tutorials for customization for your project
---
## Common Customization Parameters
>- Figure Size
>>- `fig.set_figwidth()`
>>- `fig.set_figheight()`
>- Labels
>>- ax.set_title("Name of your Title")
>>- ax.set_ylabel("Y-Label")
>>- ax.set_xlabel("X-Label)
>- `color`=\<'color'>, see [matplotlib colors](https://matplotlib.org/stable/gallery/color/named_colors.html)
>- `linestyle` = \<linestyle> see [matplotlib linestyle](https://matplotlib.org/2.0.1/api/lines_api.html#matplotlib.lines.Line2D.set_linestyle)
>- `linewidth`=\<integer>


```python
fig, ax = plt.subplots()

fig.set_figwidth(12)
fig.set_figheight(8)

ax.plot(x,y, color = 'orange', linestyle='--', linewidth=5)

fig.show()
```

```python
fig, ax = plt.subplots()

ax.set_title('A Sine Wave', fontsize = 20)

ax.set_xlabel('Time', fontsize = 10)

ax.plot(x,y, linestyle='-', marker = 'o')

fig.show()
```

# Plotting Multiple Data Series
>- Each call to `ax.plot()` will allow us to plot on the same figure
>- We can also split the figure into subplots (named *Axes*)

## Plotting on the same figure

```python
fig, ax = plt.subplots()
# Prepare data
x = np.linspace(0,10)
y1 = np.sin(x)
y2 = np.cos(x)

# Plotting two axes on the same figure
ax.plot(x,y1)
ax.plot(x,y2)

fig.show()
```

## Plotting Subplots

```python
# The values passed to subplots represent the rows, columns of the subplots
fig, (ax1,ax2) = plt.subplots(2,1)
ax1.plot(x,y1,color='blue')
ax2.plot(x,y1,color='red')
# Plot on the two axes

fig.show()
```

Another Example

```python
fig, (ax1,ax2) = plt.subplots(1,2)

fig.set_figwidth(15)

ax1.plot(x,y1,color='black')
ax2.plot(x,y1,color='gold')

fig.show()
```

# Embedding smaller charts within larger charts
We can also call define our figure object with `plt.figure()` and then add our axes with the `add_axes()` method called on the figure object to add various plots
>- `add_axes()` has a *rect* parameter which defines the dimensions `[left, bottom, width, height]` of the Axes object. All of the quantities passed are fractions (0-1) of the main figure width and height
>>- left: `0` is all the way to the left, `1` is all the way to the right
>>- Bottom: `0` is at the bottom, `1` is at the top
>>- Width and height set the size of the plot as a fraction of the figure where `1` and `1` for both width and height would be the same size as the figure.
>- See the [add_axes() doc](https://matplotlib.org/3.5.0/api/figure_api.html#matplotlib.figure.Figure.add_axes)

```python
fig = plt.figure()


# Larger figure
ax1 = fig.add_axes([0,0,1,1])
ax1.set_xlabel('Large X')
ax1.set_ylabel('Large Y')

# Smaller figure
# Change left, bottom, width, height to see how this works
ax2 = fig.add_axes([0.65,0.1,.25,.25])
ax2.set_xlabel('Small X')
ax2.set_ylabel('Small Y')

ax1.plot(x,y1,color='black')
ax2.plot(x,y1,color='gold')

fig.savefig('myfig.png', dpi=3000)

plt.show()
```
